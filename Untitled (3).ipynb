```python
import pandas as pd
import sqlite3

# Veri setini yüklüyoruz (Ayırıcı olarak ';' kullanıyoruz)
df = pd.read_csv('cardio_train.csv', sep=';')

# Verinin boyutunu kontrol edelim (Makalede 70,000 kişi olduğu belirtiliyor)
print("Veri Seti Satır/Sütun Sayısı:", df.shape)

# İlk 5 satıra göz atalım
df.head()
```

```python
# Yaşı yıla çeviriyoruz
df['age'] = (df['age'] / 365).astype(int)

# Sütun isimlerini makaleye uygun şekilde güncelliyoruz
df = df.rename(columns={
    'active': 'physically_active',
    'cardio': 'risk_of_heart_disease'
})

# Analiz için gerekmeyen 'id' sütununu çıkarıyoruz
df = df.drop('id', axis=1)

print("Ön işleme tamamlandı. Yeni sütunlar:")
print(df.columns)
df.head()
```

```python
# SQLite bağlantısı kuruyoruz (dosya yoksa otomatik oluşturulur)
conn = sqlite3.connect('kalp_verisi.db')

# Veriyi 'raw_data' adıyla tablo olarak kaydediyoruz
df.to_sql('raw_data', conn, if_exists='replace', index=False)

print("Veri 'kalp_verisi.db' içerisindeki 'raw_data' tablosuna başarıyla kaydedildi.")
```

```python
def apply_fuzzy_logic(row):
    x = row['smoke']
    y = row['alco']
    # Aktiflik 1 ise inaktivite 0'dır, Aktiflik 0 ise inaktivite 1'dir.
    not_z = 1 if row['physically_active'] == 0 else 0
    
    # Hepsi aynıysa (hepsi 0 veya hepsi 1) o değeri döndür
    if x == y == not_z:
        return float(x)
    # Herhangi biri farklıysa belirsizliği temsil eden 0.5 döndür
    else:
        return 0.5

# Yeni 'other_factors' sütununu oluşturuyoruz
df['other_factors'] = df.apply(apply_fuzzy_logic, axis=1)

# Sonucu kontrol edelim
print("Bulanıklaştırma işlemi tamamlandı.")
df[['smoke', 'alco', 'physically_active', 'other_factors']].head(10)
```

```python
def determine_fuzzy_output(row):
    # Makale Tablo 13'teki 12 kuralın Python'a aktarımı
    age = row['age']
    gender = row['gender'] # 1: Kadın, 2: Erkek
    of = row['other_factors']
    
    # KADINLAR İÇİN KURALLAR (Kural 1-6)
    if gender == 1:
        if age < 55:
            if of == 0: return 0.0     # Kural 1: Risk Yok
            if of == 0.5: return 0.5   # Kural 2: Risk Olabilir
            if of == 1: return 1.0     # Kural 3: Risk Var
        else: # age >= 55
            if of == 0: return 0.5     # Kural 4: Risk Olabilir
            if of == 0.5: return 0.5   # Kural 5: Risk Olabilir
            if of == 1: return 1.0     # Kural 6: Risk Var
            
    # ERKEKLER İÇİN KURALLAR (Kural 7-12)
    else:
        if age < 45:
            if of == 0: return 0.0     # Kural 7: Risk Yok
            if of == 0.5: return 0.5   # Kural 8: Risk Olabilir
            if of == 1: return 1.0     # Kural 9: Risk Var
        else: # age >= 45
            if of == 0: return 0.5     # Kural 10: Risk Olabilir
            if of == 0.5: return 0.5   # Kural 11: Risk Olabilir
            if of == 1: return 1.0     # Kural 12: Risk Var
    return 0.5 # Default

# Yeni hedef değişkenimizi oluşturuyoruz
df['fuzzy_target'] = df.apply(determine_fuzzy_output, axis=1)

# Makine öğrenmesi için 0.5 değerini bir sınıf olarak kodlayalım (Örn: 2)
# Makalede 0.5, '2' olarak encode edilmiştir.
df['fuzzy_target_encoded'] = df['fuzzy_target'].map({0.0: 0, 1.0: 1, 0.5: 2})

print("Kurallar uygulandı. Sınıf dağılımı:")
print(df['fuzzy_target'].value_counts())
```

```python
import time
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Modellerin listesi
models = {
    "GNB": GaussianNB(),
    "DT": DecisionTreeClassifier(max_depth=10),
    "RF": RandomForestClassifier(n_estimators=25),
    "AdaBoost": AdaBoostClassifier(n_estimators=100),
    "GB": GradientBoostingClassifier(n_estimators=90),
    "KNN": KNeighborsClassifier(n_neighbors=15),
    "SVM": SVC(kernel='linear') # SVM büyük veride yavaştır, test için gerekirse çıkarılabilir
}

# 1. Senaryo: Orijinal Veri (Makale Bölüm 2.2)
X_orig = df.drop(['risk_of_heart_disease', 'other_factors', 'fuzzy_target', 'fuzzy_target_encoded'], axis=1)
y_orig = df['risk_of_heart_disease']

# 2. Senaryo: Modifiye Veri (Makale Bölüm 2.3)
# Makaleye göre smoke, alco ve active yerine 'other_factors' kullanılır
X_mod = df[['gender', 'age', 'other_factors']]
y_mod = df['fuzzy_target_encoded']

# Sonuçları tutmak için
results = []

# Sadece GNB ve DT ile hızlı bir test yapalım (Hepsini çalıştırmak zaman alabilir)
for name, model in models.items():
    # Normal Veri Seti Eğitimi
    X_train, X_test, y_train, y_test = train_test_split(X_orig, y_orig, test_size=0.2, random_state=42)
    start = time.time()
    model.fit(X_train, y_train)
    acc_norm = accuracy_score(y_test, model.predict(X_test))
    time_norm = time.time() - start
    
    # Modifiye Veri Seti Eğitimi
    X_train, X_test, y_train, y_test = train_test_split(X_mod, y_mod, test_size=0.2, random_state=42)
    start = time.time()
    model.fit(X_train, y_train)
    acc_mod = accuracy_score(y_test, model.predict(X_test))
    time_mod = time.time() - start
    
    results.append([name, acc_norm, acc_mod, time_norm, time_mod])

# Sonuçları tabloya dökme
results_df = pd.DataFrame(results, columns=['Model', 'Normal Acc', 'Fuzzy Acc', 'Normal Time', 'Fuzzy Time'])
results_df.to_sql('model_results', conn, if_exists='replace', index=False)
results_df
```
